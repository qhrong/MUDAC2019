```python
import pandas as pd
import numpy as np
```

```python
orig_data = pd.read_csv('Train_July_X10_Y0_v1.csv')
```

```python
import pywt
```

```python
price = orig_data['price']
```

```python
(ca, cd) = pywt.dwt(price, "haar")
```

```python
cat = pywt.threshold(ca, np.std(ca), mode="soft")
cdt = pywt.threshold(cd, np.std(cd), mode="soft")
```

```python
# Inverse discret wavelet transform
tx = pywt.idwt(cat, cdt, "haar")
```

```python
tx = tx[1:]
```

```python
pd.DataFrame(tx).to_csv('wavelet_May.csv')
```

```python
import matplotlib.pyplot as plt
```

```python
fig = plt.figure()
ax = plt.axes()
fig= plt.figure(figsize=(6,3))
ax.plot(tx,linewidth=0.5)
ax.plot(orig_data['price'],linewidth=0.5, color='r')
```
